```python
df = pd.read_csv('/content/drive/MyDrive/comm_med_gfr/kidney_transplantation_work_pipeline_17_05_2023/Multilinear_Regression_new_021024/dat_red_new.csv')
# Select the first column
first_col = df.iloc[:, 2]

# Select the two columns before the last one
two_before_last_cols = df.iloc[:, -3:-1]

# Combine these into X
X = pd.concat([first_col, two_before_last_cols], axis=1).values
#X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
X
```

```python
from sklearn import linear_model
import statsmodels.api as sm
from scipy import stats
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```
